```python
import tevatron
from tevatron.modeling import DenseModelForInference
from transformers import AutoConfig, AutoTokenizer
import torch
```

```python
tevatron.__file__
```

```python
model_name_or_path = './co-condenser-marco/'
num_labels = 1
```

```python
config = AutoConfig.from_pretrained(
    model_name_or_path,
    num_labels=num_labels,
    cache_dir=model_name_or_path,
)
tokenizer = AutoTokenizer.from_pretrained(
    model_name_or_path,
    cache_dir=model_name_or_path
)

model = DenseModelForInference.build(
    model_name_or_path,
    config=config,
    cache_dir=model_name_or_path,
)
```

```python
for name, param in model.named_parameters():
    print(name, param.size())
```

```python
len(list(model.named_parameters())), len(list(model.parameters()))
```

```python
x = torch.rand([1, 768, 768])
```

```python
eq = model.lm_q.encoder(x)
```

```python
ep = model.lm_p.encoder(x)
```

```python
torch.allclose(eq.last_hidden_state, ep.last_hidden_state)
```
